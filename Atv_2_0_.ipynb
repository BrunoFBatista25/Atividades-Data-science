```python
import matplotlib.pyplot as plt
import pandas as pd

Df = pd.read_excel(r'/content/default_of_credit_card_clients__courseware_version_1_21_19 (1).xlsx')

Faturas = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']
Pagamentos = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
pd.set_option('display.float_format', '{:.2f}'.format)
```

# Resposta 2

```python
Dados = Df[Faturas].describe()
print(Dados,"\n")
```


Ao usar o método '.describe()' nos deparamos com certas informações estátisticas dos dados das colunas financeiras do Dataframe.
* Count: conta quantidade das linhas, ou seja clientes do banco.

* Max e Min:
 Os máximos e mininos no contexto financeiro fornecem o valor minimo e o máximo do saldo devedor, quando o valor minino é negativo temos, na verdade, crédito do cliente.

* Std e média:
 std Constitue o gráfico de dispersão da distribuição enquanto a média se encontra como média da distribuição. É como o std se encontra maior que a médio podemos afirmar que temos uma relação assimetrica e temos muitos outliers no dados, como também na análise dos interquatilicos(25,50,75) propõe.


#Resposta 3

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('visualização das Faturas', fontsize = 20)
axs = axs.flatten()

for i,Fatura in enumerate(Faturas):
  Desc = Df[Fatura].describe()
  Max_Df = Desc.loc['max']
  Min_Df = Desc.loc['min']

  axs[i].hist(Df[Fatura], bins = 20,color='skyblue', edgecolor='black');
  axs[i].set_title(f'{Fatura}')
  axs[i].grid(True,alpha=0.8)
  axs[i].set_xlim(Min_Df, Max_Df)


plt.tight_layout()
plt.show()
```

Visualização dos dados das Faturas:

Para a visualização do histograma foram pegos os o menor valor do gráfico e maior valor.

Os histogramas revelam que o banco foi reduzindo suas dividas e com o passar do tempo os valores nulos dentro das listas eles não representam mais uma parte significativa do total das Faturas, o que revela o banco nestes 6 meses ainda há de receber dos juros dos emprestimos ainda não pagados futuramente.

#Resposta 5

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('visualização dos Pagamentos', fontsize = 20)
axs = axs.flatten()
for i,Pagamento in enumerate(Pagamentos):
  Desc = Df[Pagamento].describe()
  Max_Df = Desc.loc['max']
  Min_Df = Desc.loc['min']
  axs[i].hist(Df[Pagamento], bins = 20,color='skyblue', edgecolor='black');
  axs[i].set_title(f'{Pagamentos[i]}')
  axs[i].grid(True,alpha=0.8)
  axs[i].set_xlim(-8000, 100000)
  axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=30)

plt.tight_layout()
plt.show()
```

#Resposta 7

```python
import numpy as np
Máscara_pagamento2 = (Df[Pagamentos] != 0)

log_pag = Df[Pagamentos][Máscara_pagamento2].apply(np.log10)

fig, axs = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('Pagamentos na base logaritimica', fontsize=12)
axs = axs.flatten()

for i, v in enumerate(Pagamentos):
    axs[i].hist(log_pag[v].dropna(), bins=20, color='skyblue', edgecolor='black')
    axs[i].set_title(f'log10({v})')
    axs[i].grid(True, alpha=0.8)
    axs[i].set_xlabel('log10')
    axs[i].set_ylabel('Frequência')
    axs[i].set_xlim(0, 6)

plt.tight_layout()
plt.show()

```
