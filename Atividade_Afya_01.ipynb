```python
#Resposta 3 | Dados numéricos e Discretos
Dados_Discretos = ('ID','SEX','EDUCATION','MARRIAGE','PAY_1','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6','default payment next month')
Dados_Numéricos = ('LIMIT_BAL','AGE','BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6','PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6')

# Dados Numéricos
print("\n--- Intervalos de valores de dados numéricos---")
for col in Dados_Numéricos:
    if col in Df.columns:
        minimo = Df[col].min()
        maximo = Df[col].max()
        print(f"{col}: mínimo = {minimo}, máximo = {maximo}")
    else:
        print(f"{col}: COLUNA NÃO ENCONTRADA")

# Dados Discretos
print("\n--- Frequência de valores  de dados discretos ---")
for col in Dados_Discretos:
    if col in Df.columns:
        print(f"\n{col} - Valores únicos e suas contagens:")
        print(Df[col].astype(str).value_counts().sort_index())
    else:
        print(f"{col}: COLUNA NÃO ENCONTRADA")
```
